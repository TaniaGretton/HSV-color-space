```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
```

```python
image = cv2.imread('/Users/taniagretton/miniconda3/CHALLENGES/water balloons.jpeg')
image = cv2.cvtColor(image,cv2.COLOR_BGR2RGB)
plt.imshow(image)
```

# Plot color channels

```python
r = image[:,:,0]
g = image[:,:,1]
b = image[:,:,2]

f,(ax1,ax2,ax3) = plt.subplots(1,3,figsize=(20,10))
ax1.set_title('Red')
ax1.imshow(r, cmap='gray')

ax2.set_title('Green')
ax2.imshow(g,cmap='gray')

ax3.set_title('Blue')
ax3.imshow(b,cmap='gray')
```

```python
# convert from RGB to HSV
hsv = cv2.cvtColor(image,cv2.COLOR_RGB2HSV)

# HSV channels
h = hsv[:,:,0]
s = hsv[:,:,1]
v = hsv[:,:,2]

# plot HSV channels
f,(ax1,ax2,ax3) = plt.subplots(1,3,figsize=(20,10))
ax1.set_title('Hue')
ax1.imshow(h,cmap='gray')

ax2.set_title('Saturation')
ax2.imshow(s,cmap='gray')

ax3.set_title('Value')
ax3.imshow(v,cmap='gray')
```

## Define pink and hue selection treshhold

```python
# in RGB values
lower_pink = np.array([180,0,100])
upper_pink = np.array([255,255,230])

# in HSV values
lower_hue = np.array([160,0,0])
upper_hue = np.array([180,255,255])
```

### Mask the image for RGB

```python
mask_rgb = cv2.inRange(image,lower_pink,upper_pink)
masked_rgb = np.copy(image)
masked_rgb[mask_rgb==0] = [0,0,0]
plt.imshow(masked_rgb)
```

### MAsk the image for HSV

```python
mask_hsv = cv2.inRange(hsv,lower_hue,upper_hue)
masked_hsv = np.copy(image)
masked_hsv[mask_hsv==0] = [0,0,0]
plt.imshow(masked_hsv)
```
